```python
#Drop ID column
df2 = df.drop(columns=['Passenger ID'], inplace=False)
```

```python
df2['Arrival Airport'] = df2['Arrival Airport'].apply(lambda x: np.nan if x == '0' else x)
df3 = df2.dropna(subset=['Arrival Airport'])
```

```python
df3.info()
```

```python
#Replace departure date anomaly
df3['Departure Date'] = df3['Departure Date'].str.replace('-','/')
```

```python
#Membuat kolom baru khusus month 

df3['Departure Date'] = pd.to_datetime(df3['Departure Date'], format='%m/%d/%Y')
df3['Month'] = df3['Departure Date'].dt.month_name()
```
